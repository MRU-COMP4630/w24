```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def P(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-x**2/2)

x_s = np.random.randn(1000)
plt.hist(x_s, density=True, bins=30)
x = np.linspace(-3, 3, 100)
plt.plot(x, P(x), 'r')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.savefig('standard_normal_distribution.png')
```

```python
def E(x, f):
    return P(x) @ f(x) * (x[1] - x[0]) # continuous case

# Expected value for f(x) = x
print(E(x, lambda x: x))

# Expected value for f(x) = x^2
print(E(x, lambda x: x**2))
```

```python
# Information
plt.plot(x, -np.log(P(x)))
plt.xlabel('x ~ N(0, 1)')
plt.ylabel('I(x)')
plt.savefig('standard_normal_information.png')
```

```python
# Bernoulli entropy
p = np.linspace(0, 1, 100)
plt.plot(p, -p*np.log(p) - (1-p)*np.log(1-p))
plt.xlabel('p')
plt.ylabel('H(p)')
plt.savefig('bernoulli_entropy.png')
```

```python
# Cross-entropy loss for binary classification
p = np.linspace(0.01, 0.99, 100)
y = 0
plt.plot(p, -y*np.log(p) - (1-y)*np.log(1-p), label='$y=0$')
y = 1
plt.plot(p, -y*np.log(p) - (1-y)*np.log(1-p), label='$y=1$')
plt.xlabel('$p$')
plt.ylabel('$\mathcal{L}(y, p)$')
plt.legend()
plt.savefig('cross_entropy_loss.png')
```
